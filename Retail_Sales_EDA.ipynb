```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/sample_data/Superstore.csv', encoding='latin1')

df.head()

df.shape

df.columns

df.info()

df.describe()

df.isnull().sum()

df.drop_duplicates(inplace=True)

df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()

df.groupby('Category')['Profit'].sum().plot(kind='bar', color='green')
plt.title("Profit by Category")
plt.show()

df.groupby('Sub-Category')['Sales'].sum().sort_values().plot(kind='barh', figsize=(10,6))
plt.title("Sales by Sub-Category")
plt.show()

df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.show()

month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

df.groupby('Month')['Sales'].sum().plot(kind='bar')
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.show()

loss_df = df[df['Profit'] < 0]

loss_df.groupby('Sub-Category')['Profit'].sum().plot(kind='bar', color='red')
plt.title("Loss Making Sub-Categories")
plt.show()

plt.figure(figsize=(8,5))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()

df.to_csv("Processed_Retail_Data.csv", index=False)
print("Saved successfully!")
```

# 📊 Key Insights

- Technology category generates the highest profit
- Some sub-categories result in losses
- Sales vary across states significantly
- Monthly sales show seasonal variation
- Profit is not always directly proportional to sales
